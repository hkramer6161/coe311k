```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 0, 1, 0, 1])

cs = CubicSpline(x, y, bc_type='natural')
 
coefficients = np.polyfit(x, y, 3)
polynomial = np.poly1d(coefficients)

x_interp = np.linspace(0, 5, 100)
y_interp_cs = cs(x_interp)
y_interp_poly = polynomial(x_interp)

plt.figure()
plt.plot(x, y)
plt.plot(x_interp, y_interp_poly, label='Poly')
plt.plot(x_interp, y_interp_cs, label='Cs')
plt.legend()
plt.show()
```

The cubicspline fit is far better than the polyfit and lies much closer to the data points when the polyfit is set to degree 3.
