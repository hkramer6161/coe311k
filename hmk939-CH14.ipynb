```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: np.sin(x)
a, b = 0, 2 * np.pi # Interval
N = 50 # Number of sample points

# Create a uniform grid
x = np.linspace(a, b, N)
h = (b - a) / (N - 1)

exact_deriv = np.cos(x)
forward_diff = (f(x + h) - f(x)) / h
backward_diff = (f(x) - f(x - h)) / h
central_diff = (f(x + h) - f(x - h)) / (2 * h)
```

Calculate all FDM and make grid

```python
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.plot(x, exact_deriv)
plt.title("Actual derivative")

plt.subplot(2, 2, 2)
plt.plot(x, forward_diff)
plt.title("Forward difference")

plt.subplot(2, 2, 3)
plt.plot(x, backward_diff)
plt.title("Backward difference")

plt.subplot(2, 2, 4)
plt.plot(x, central_diff)
plt.title("Central difference")

plt.show()
```

the plots show that the central diff method and backward diff method are closest to the actual derivative and the forward diff method is a bit skewed to the left.
