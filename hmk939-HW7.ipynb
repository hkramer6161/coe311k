```python
import numpy as np
import matplotlib.pyplot as plt


x_start = 0
x_end = 10 * np.pi
y_0 = 1.0
h_values = [0.01, 0.05, 0.15, 0.2, 0.25, 0.5, 1.0]

def analytical_solution(x):
    return np.exp(-2 * np.sin(x))
```

define parameters and create function for the analytical solution for the euation

```python
def euler_method(h, x_start, x_end, y_0):
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))

    y_values[0] = y_0

    for i in range(1, len(x_values)):
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]

        y_values[i] = y_n + h * (-2 * y_n * np.cos(x_n))

    return x_values, y_values
```

euler method function

```python
avg_errors = []
local_errors = []

plt.figure()

for h in h_values:
    x, y = euler_method(h, x_start, x_end, y_0)
    y_actual = analytical_solution(x)

    avg_errors.append(np.mean(np.abs(y - y_actual)))

    idx = np.argmin(np.abs(x - 15))
    local_errors.append(abs(y[idx] - analytical_solution(x[idx])))

    plt.plot(x, y, label=f"h={h}")
    
plt.xlabel("x")
plt.ylabel("y")
plt.title("Euler's Method for varrying h values")
plt.legend()
plt.grid(True)
plt.show()

plt.figure()
plt.plot(h_values, avg_errors)
plt.xlabel("h")
plt.ylabel("Average Error")
plt.title("Average Error vs h")
plt.grid(True)
plt.show()

plt.figure()
plt.plot(h_values, local_errors)
plt.xlabel("h")
plt.ylabel("Local error around x = 15")
plt.title("Local error vs h")
plt.grid(True)
plt.show()
```

As the h value increases, the euler method deviates further from the actual solution. Around x=15 this is showed well as it increases across all h values and doesnt decrease at an h=1 like in the average error plot. In the average error plot this occurs just because the h = 1 step size gets "lucky".
