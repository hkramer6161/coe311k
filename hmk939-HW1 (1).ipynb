```python
import numpy as np
import matplotlib.pyplot as plt
```

Header

```python
def approximatePi(numSamples):
    counter = 0
    for s in range(numSamples):
        x = np.random.uniform(-1, 1)
        y = np.random.uniform(-1, 1)
        
        if (x*x + y*y < 1):
            counter += 1
    
    return 4*counter/numSamples
            
```

Montecarlo pi function

```python
np.random.seed(0)

iteration = np.arange(10, 10000, 20)
num_iteration = len(iteration)
monteOut = np.empty(num_iteration)
error = np.empty(num_iteration)
count = -1
```

initiate random, get number of iterations, make arrays, and make counter

```python
for n in iteration:
    count += 1
    monteOut[count] = approximatePi(n)
    error[count] = np.round(abs(monteOut[count] - np.pi), 15)
```

for loop that makes array of monte carlo output and error values rounded

```python
plt.plot(iteration, error)
plt.xlabel('number of samples')
plt.ylabel('error')
plt.show()
```

plots number of samples vs the error and labels axis
