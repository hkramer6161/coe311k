```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters

k = 1.0        # Constant in the equation (e.g., spring constant)
h = 0.01       # Step size
x_start = 0    # Start of the interval
x_end = 10     # End of the interval
y_0 = 1.0      # Initial condition for y
v_0 = 0.0      # Initial condition for dy/dx
h_values = [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]
k_values = [0.1, 1, 5, 10, 50]
```

```python
# Euler's Method loop
plt.figure(figsize=(10, 5))

for h in h_values:
    # Create arrays to store x, y, and v values
    
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))
    
    # Set initial conditions
    
    y_values[0] = y_0
    v_values[0] = v_0
    
    for i in range(1, len(x_values)):
        # Current values
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]    

        # Update equations
        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)
        
    plt.plot(x_values, y_values, label=f'Position y(x), h = {h}')
    
# Plot the results

plt.plot(x_values, v_values, label='Velocity v(x)', linestyle='--')
plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method for differing h values')
plt.legend()
plt.grid(True)

plt.show()
```

as the h value increases, so does the instability. Its very clearly unstable around an h of .1 or .2. 

```python
plt.figure(figsize=(10, 5))

for k in k_values:
    h= .01
    x1_values = np.arange(x_start, x_end + h, h)
    y1_values = np.zeros(len(x1_values))
    v1_values = np.zeros(len(x1_values))

    y1_values[0] = y_0
    v1_values[0] = v_0

    for i in range(1, len(x1_values)):
        y_n1 = y1_values[i - 1]
        v_n1 = v1_values[i - 1]

        y1_values[i] = y_n1 + h * v_n1
        v1_values[i] = v_n1 - h * k * y_n1

    plt.plot(x1_values, y1_values, label=f'Position y(x), k = {k}')

plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method for differing k values')
plt.legend()
plt.grid(True)

plt.show()
```

as the k value increases, the amplitude increases over x. At k =10 h = .01 the solution remains stable for longer. At larger h values though the euler method would not be very effective for solving it.
