```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
bfl1 = m * X + c

print(f"Slope: {m}, Intercept: {c}")

plt.figure()
plt.plot(X, Y)
plt.plot(X, bfl1)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Part 1")
plt.show()
```

First part, plots data and best fit line using least squares

```python
np.random.seed(0)     

X2 = np.sort(np.random.rand(10) * 10) 
Y2 = np.random.rand(10) * 10   

A2 = np.vstack([X2, np.ones(len(X2))]).T
m2, c2 = np.linalg.lstsq(A2, Y2, rcond=None)[0]
bfl2 = m2 * X2 + c2

print(f"Slope: {m2}, Intercept: {c2}")

plt.figure()
plt.plot(X2, Y2)
plt.plot(X2, bfl2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Part 2")
plt.show()
```
