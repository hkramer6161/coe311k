```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

header for imports

```python
def sin_taylor_series(x, numTerms):
    total = 0
    for k in range(numTerms):
        term = (((-1)**k)*(x**(2*k+1))/math.factorial(2*k+1))
        total += term
    return total
```

function that takes in x value and number of taylor series terms and returns taylor sin approximation at that x value with that number of terms

```python
taylor_out_1 = []
taylor_out_2 = []
taylor_out_3 = []
taylor_out_4 = []
error_1 = []
error_2 = []
error_3 = []
error_4 = []
taylor_in = np.arange(0, 10.1, .1)
```

make arrays to store taylor series approximations for 1 to 4 terms and error. Counter is for ensuring 101 values in array.

```python
for i in range(len(taylor_in)):
    term1 = sin_taylor_series(taylor_in[i], 1)
    term2 = sin_taylor_series(taylor_in[i], 2)
    term3 = sin_taylor_series(taylor_in[i], 3)
    term4 = sin_taylor_series(taylor_in[i], 4)
    taylor_out_1.append(term1)
    taylor_out_2.append(term2)
    taylor_out_3.append(term3)
    taylor_out_4.append(term4)
    error_1.append(abs(term1 - np.sin(taylor_in[i])))
    error_2.append(abs(term2 - np.sin(taylor_in[i])))
    error_3.append(abs(term3 - np.sin(taylor_in[i])))
    error_4.append(abs(term4 - np.sin(taylor_in[i])))
```

for loop that stores the taylor approximations as arrays then calculates the errors as arrays

```python
plt.plot(taylor_in, error_1, label="1 term")
plt.plot(taylor_in, error_2, label="2 term")
plt.plot(taylor_in, error_3, label="3 term")
plt.plot(taylor_in, error_4, label="4 term")
plt.title("taylor series error vs number of terms")
plt.legend()
plt.show()
```

plot error vs number of taylor series terms

as you include more terms the error is less near 0 but increases quicker with more terms as you get away from 0. For small values of x more terms is more accurate but for larger values its less accurate.
