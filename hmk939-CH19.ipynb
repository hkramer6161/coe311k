```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
h_values = [0.001]
t_start = 0
t_end = 30
y_0 = 0.0

# Exact solution
def analytical_solution(t):
    return 3 - 0.998*np.exp(-1000*t) - 2.002*np.exp(-t)
```

```python
def euler_forward(h, t_start, t_end, y_0):
    t_values = np.arange(t_start, t_end + h, h)
    y_values = np.zeros(len(t_values))

    y_values[0] = y_0

    for i in range(1, len(t_values)):
        t_n = t_values[i - 1]
        y_n = y_values[i - 1]

        y_values[i] = y_n + h * (-1000*y_n + 3000 - 2000*np.exp(-t_n))

    return t_values, y_values
```

```python
t_analytical = np.linspace(t_start, t_end, 5000)
y_analytical = analytical_solution(t_analytical)

plt.figure(figsize=(12, 6))
plt.plot(t_analytical, y_analytical, label='Analytical solution')

for h in h_values:
    t, y = euler_forward(h, t_start, t_end, y_0)
    plt.plot(t, y, label=f"Euler Forward, h = {h}")

plt.xlabel("t")
plt.ylabel("y")
plt.title("Euler Forward vs Analytical Solution")
plt.legend()
plt.grid(True)

plt.show()
```

the method fails for an h of .1 and .01 because the values get too big to be stored as floating points. For .001 eulers forward is very close to the analytical solution.
