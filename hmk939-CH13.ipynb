```python
import numpy as np
import matplotlib.pyplot as plt
import time

def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]

    return C
```

Naive multiplication function

```python
sizes = [3, 6, 12, 24, 48, 96, 192, 384]
naive_times = []
numpy_times = []

for n in sizes:
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)

    start = time.time()
    naive_matrix_multiplication(A.tolist(), B.tolist())
    end = time.time()
    naive_time = (end - start)
    naive_times.append(naive_time)

    start = time.time()
    np.dot(A, B)
    end = time.time()
    numpy_time = (end - start)
    numpy_times.append(numpy_time)
    
    print("Naive time for matrix size ", n, " ", naive_time)
    print("Numpy time for matrix size ", n, " ", numpy_time)
```

run both functions for different matrix sizes and collect times, time are much longer for naive matrix multiplication

```python
plt.plot(range(1, 9), naive_times)
plt.xlabel("Matrix Number")
plt.ylabel("Runtime")
plt.title("Naive")
plt.show()

plt.plot(range(1, 9), numpy_times)
plt.xlabel("Matrix Number")
plt.ylabel("Runtime")
plt.title("NumPy")
plt.show()
```

The plots show that the naive matrix multiplication function increases in runtime more rapidly as the matrix size increases than the numpy function.
