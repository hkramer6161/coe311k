```python
import numpy as np
import matplotlib.pyplot as plt

x_start, x_end = 0, 2
h_vals = list(np.arange(0.01, 1.0, 0.05))
h_vals.append(1.0)

# Function representing dy/dx = -2y
def f(x, y):
    return -2 * y
```

import libraries and make function

```python
for h in h_vals:
    num_steps = int((x_end - x_start) / h)

    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)

    # Initial condition
    y_values[0] = 1  # y(0) = 1

    # Implement Euler's method
    for i in range(num_steps):
        step = x_values[i + 1] - x_values[i]
        y_values[i + 1] = y_values[i] + step * f(x_values[i], y_values[i])

    # Plot the results
    plt.plot(x_values, y_values, label=f"h = {h:.2f}")

    if h == .01:
        first_x = x_values
        first_y = y_values
        print("Run h = 0.01")
        for i in range(len(x_values)):
            print(f"x = {x_values[i]:.2f}, error = 0")
        print("Total accumulated error = 0")
    else:
        first_y_val = np.interp(x_values, first_x, first_y)
        errors = np.abs(y_values - first_y_val)
        total_error = np.sum(errors)

        print("\n Run h = ", h)
        for i in range(len(x_values)):
            print(f"x = {x_values[i]:.2f}, error = ", errors[i])
        print("Total accumulated error = ", total_error)
        
plt.xlabel("x")
plt.ylabel("y")
plt.title("Approximation of dy/dx = -2y using Euler's Method")
plt.legend()
plt.show()
```

as the step size increases, the total error also increases.
