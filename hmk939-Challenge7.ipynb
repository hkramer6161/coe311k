```python
import numpy as np
import time

def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]

        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new

    return x, max_iterations
```

Jacobi method function

```python
def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    
    for k in range(max_iterations):
        x_new = x.copy()
        
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
            
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        
        x = x_new
        
    return x, max_iterations
```

Gauss-sidel function

```python
A = np.array([[0.582745, 0.48, 0.10, 0, 0], [0.48, 1.044129, 0.46, 0.10, 0], [0.10, 0.46, 1.10431, 0.44, 0.10], [0, 0.10, 0.44, 0.963889, 0.42], [0, 0, 0.10, 0.42, 0.522565]], dtype=float)
b = np.array([1.162745, 2.084129, 2.20431, 1.923889, 1.042565], dtype=float)
A1 = np.array([[1, 1], [1, 1.0001]], dtype=float)
b1 = np.array([1, 1], dtype=float)
A2 = np.array([[1, 1], [1, 1.0001]], dtype=float)
b2 = np.array([1, 1.0001], dtype=float)
x0 = np.zeros_like(b)
x01 = np.zeros_like(b1)
x02 = np.zeros_like(b2)
tol = 1e-6
max_iterations = 1000000
tij = time.perf_counter()
solutionj, iterationsj = jacobi(A, b, x0, tol, max_iterations)
tj = time.perf_counter() - tij
ti = time.perf_counter()
solution, iterations = gauss_seidel(A, b, x0, tol, max_iterations)
t = time.perf_counter() - ti
print(f"Solution Jacobi part a: {solutionj}")
print(f"Iterations Jacobi part a: {iterationsj}")
print(f"Time Jacobi part a: {tj}")
print(f"Solution Gauss part a: {solution}")
print(f"Iterations Gauss part a: {iterations}")
print(f"Time Gauss part a: {t}")
tij1 = time.perf_counter()
solutionj1, iterationsj1 = jacobi(A1, b1, x01, tol, max_iterations)
tj1 = time.perf_counter() - tij1
ti1 = time.perf_counter()
solution1, iterations1 = gauss_seidel(A1, b1, x01, tol, max_iterations)
t1 = time.perf_counter() - ti1
print(f"Solution Jacobi part b: {solutionj1}")
print(f"Iterations Jacobi part b: {iterationsj1}")
print(f"Time Jacobi part b: {tj1}")
print(f"Solution Gauss part b: {solution1}")
print(f"Iterations Gauss part b: {iterations1}")
print(f"Time Gauss part b: {t1}")
tij2 = time.perf_counter()
solutionj2, iterationsj2 = jacobi(A2, b2, x02, tol, max_iterations)
tj2 = time.perf_counter() - tij2
ti2 = time.perf_counter()
solution2, iterations2 = gauss_seidel(A2, b2, x02, tol, max_iterations)
t2 = time.perf_counter() - ti2
print(f"Solution Jacobi part c: {solutionj2}")
print(f"Iterations Jacobi part c: {iterationsj2}")
print(f"Time Jacobi part c: {tj2}")
print(f"Solution Gauss part c: {solution2}")
print(f"Iterations Gauss part c: {iterations2}")
print(f"Time Gauss part c: {t2}")
```
